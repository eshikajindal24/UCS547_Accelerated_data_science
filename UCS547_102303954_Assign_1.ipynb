3. Find the most frequent value and their indice(s) in the following arrays:
i. x = np.array([1,2,3,4,5,1,2,1,1,1])
ii. y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, 3])

```python
import numpy as np
x = np.array([1,2,3,4,5,1,2,1,1,1])
frequency = np.bincount(x)
most_freq = np.argmax(frequency)
i = np.where(x==most_freq)
print("Freq - ",most_freq)
print("Indices - ",i)

y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, 3])
freq = np.bincount(y)
most_freq_y = np.argmax(freq)
j = np.where(y==most_freq_y)
print("Freq - ",most_freq_y)
print("Indices - ",j)
```

4. For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'),
find i. Sum of all elements
ii. Sum of all element’s row-wise
iii. Sum of all element’s column-wise.

```python
import numpy as np
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
# (i)
total = np.sum(gfg)
print(total)

# (ii)
total_row = np.sum(gfg, axis=1)
print(total_row)

# (iii)
total_col = np.sum(gfg, axis=0)
print(total_col)
```
